# One Sample T-test



> TIP: shit + tab to help get the documentation, tab to get methods

```python
# create a 1d array from a normal dist 0/1
# size = 15
x=np.random.normal(size=15)
x
```

```python
x.shape
```

```python
x.std()
```

```python
x.mean()
```

```python
# ttest from scipy
stats.ttest_1samp(x,0)
```

```python
# try this again - larger sample
# size = 100
y=np.random.normal(size=100)
y
```

```python
stats.ttest_1samp(y,0)
```

> Makes sense right, as sample size increases from the distribution, pvalue gets larger.  Much less unlikely to be from a different dist

```python
# another but we shift the data to 2/1, size=50
z=np.random.normal(loc=2,scale=1,size=50)
```

```python
# save out the result to a variable
# result
result=stats.ttest_1samp(z,0)
result
```

# Two Sample t-test

```python
# lets create two random normal, 100/15, 115/15, size=100
x_1=np.random.normal(100,15,100)
y_1=np.random.normal(115,15,100)
```

```python
print(x_1.mean())
print(y_1.mean())
```

```python
stats.ttest_ind(x_1,y_1)
```

> Assumes equal variances.  Review help to set if not.

# Chi-square

```python
## test for indepdence
# example set from here
# 4 sets of rolls of dice, summarized

a1 = [6, 4, 5, 10]
a2 = [8, 5, 3, 3]
a3 = [5, 4, 8, 4]
a4 = [4, 11, 7, 13]
a5 = [5, 8, 7, 6]
a6 = [7, 3, 5, 9]
dice = np.array([a1, a2, a3, a4, a5, a6])
```

```python
dice
```

```python
stats.chi2_contingency(dice)
```

```python
# another way to unpack the results
stat, p, dof, exp=stats.chi2_contingency(dice)
```

```python
p
```

> What just happened?  We can unpack the results  by itemizing the values as variables.  You will see this quite a bit in the scikit learn syntax when doing train/test splits.

![](https://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/Figure3.PNG)

```python
# another example
# https://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData5.html
# by group, yes/no

g1 = [5, 78]
g2 = [21, 63]
data = np.array([g1, g2])
data
```
